```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.NearsidePerspective(central_longitude=3, 
                                                  central_latitude=48))
plt.title('$\it{RGF93 / CC48}$', fontweight="bold", pad=20)
ax.stock_img()
ax.gridlines(draw_labels=True)

# epsg:3948 extents
#lat_corners = np.array([618180.541, 618180.54, 6527546.03, 6527546.03])
#lon_corners = np.array([2211981.05, 7620707.95, 7620707.95, 2211981.05]) 
lat_corners = np.array([41.15, 41.15, 41.15, 51.56, 51.56, 51.56])
lon_corners = np.array([-9.86, 0.0, 10.38, 10.38, 0.0, -9.86])

poly_corners = np.zeros((len(lat_corners), 2), np.float64)
poly_corners[:,0] = lon_corners
poly_corners[:,1] = lat_corners

poly = mpatches.Polygon(poly_corners, closed=True, #ec='r', 
                        fill=True, lw=1, fc="grey",
                        alpha=0.4, 
                        #transform=ccrs.epsg(3948))
                        transform=ccrs.Geodetic(),
                        label = 'Projected bounds')

# epsg:3948 center coordinates
plt.plot(1489088.51, 7020857.23, color='blue', marker='o',
         markersize=1, transform=ccrs.epsg(3948)) 
    
ax.add_patch(poly)

lat_corners2 = np.array([47.5, 47.5, 47.5, 48.5, 48.5,  48.5])
lon_corners2 = np.array([-9, 0.0, 10, 10, 0.0, -9]) 

poly_corners2 = np.zeros((len(lat_corners2), 2), np.float64)
poly_corners2[:,0] = lon_corners2
poly_corners2[:,1] = lat_corners2

poly2 = mpatches.Polygon(poly_corners2, closed=True, ec='r', 
                        fill=False, lw=1, #fc="grey",
                        alpha=1, 
                        #transform=ccrs.epsg(2048))
                        transform=ccrs.Geodetic(),
                        label = 'RGF93 / CC48 extent')
ax.add_patch(poly2)

gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                  linewidth=2, color='grey', alpha=0.4, linestyle='--')
gl.xlabels = False
gl.ylabels = False
#gl.left_labels = True
#gl.ylines = True
gl.ylocator = mticker.FixedLocator([48])
gl.xlocator = mticker.FixedLocator([3])
gl.xformatter = LongitudeFormatter()
gl.yformatter = LatitudeFormatter()
gl.xlabel_style = {'size': 15, 'color': 'black'}
gl.ylabel_style = {'color': 'grey', 'weight': 'bold'}

#add legend
ax.legend(bbox_to_anchor=(1.2, 1.05), loc='upper right')
#save
plt.savefig('./img/RGF93_CC48.png')
plt.show()
```
